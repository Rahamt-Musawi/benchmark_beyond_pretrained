```python
import gradio as gr
from ipywidgets import FloatProgress
from ciphers import encrypt_decrypt, normalize_greek

english_alphabet = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
greek_alphabet = 'αβγδεζηθικλμνξοπρστυφχψωΑΒΓΔΕΖΗΘΙΚΛΜΝΞΟΠΡΣΤΥΦΧΨΩ'

def process_text(input_text, task, method, key, language):
    if language == "English":
        alphabet = english_alphabet
    else:  # Greek
        alphabet = greek_alphabet
        input_text = normalize_greek(input_text)
        if method == "Vigenère Cipher":
            key = normalize_greek(key)

    cipher_method = "caesar_cipher" if method == "Caesar Cipher" else "vigenere_cipher"

    try:
        if cipher_method == "caesar_cipher":
            try:
                key = int(key)
            except ValueError:
                return "Error: Caesar Cipher requires an integer key."

        result = encrypt_decrypt(cipher_method, task.lower(), input_text, key, alphabet)
        return result
    except ValueError as e:
        return str(e)

# Create the Gradio interface
iface = gr.Interface(
    fn=process_text,
    inputs=[
        gr.Textbox(label="Input Text"),
        gr.Dropdown(["Encrypt", "Decrypt"], label="Task"),
        gr.Dropdown(["Caesar Cipher", "Vigenère Cipher"], label="Method"),
        gr.Textbox(label="Key", placeholder="Integer for Caesar, Text for Vigenère"),
        gr.Dropdown(["English", "Greek"], label="Language")
    ],
    outputs=gr.Textbox(label="Result"),
    title="Caesar and Vigenère Cipher App",
    description="Encrypt or decrypt text using Caesar or Vigenère cipher for English or Greek languages.",
    examples=[
        ["Hello, World!", "Encrypt", "Caesar Cipher", "3", "English"],
        ["Khoor, Zruog!", "Decrypt", "Caesar Cipher", "3", "English"],
        ["Hello, World!", "Encrypt", "Vigenère Cipher", "KEY", "English"],
        ["Γειά σου, Κόσμε!", "Encrypt", "Caesar Cipher", "3", "Greek"],
        ["Γειά σου, Κόσμε!", "Encrypt", "Vigenère Cipher", "ΚΛΕΙΔΙ", "Greek"]
    ]
)

# Launch the app
iface.launch(share=True)
```
